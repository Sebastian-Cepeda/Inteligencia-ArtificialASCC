```python
from pickle import TRUE


FINAL_STATE =    [['1','1','1','1','a'],
                  ['2','2','2','2','b'],
                  ['3','3','3','3','c'],
                  ['4','4','4','4','d'],
                  ['5','5','5','5','e']]

#Para hacer una generacion automatica deberia de un estado inicial cualquiera
#Seria preferible una funciona que designe un lugar aleatorio para los esapcios vacios 
# e y s, les cambie de nombre para mayor comodiad en la busqueda 
#depues empezar a rellenar cada espacio vacio con la logica
#  de cada caracter 1, 2, 3 <= 5 y  4, 5 <= 4 en todas las listas

                  
INITIAL_STATE =  [['a','2','3','5','1'],
                  ['d','2','3','1','3'],
                  ['4','5','c','3','e'],
                  ['b','1','4','2','5'],
                  ['1','2','2','4','5']]

#ir e ic son las posiciones donde se encuentra el elemento 
#para no olvidarlo como funciona enumerate una estructura que pueda recorrer
# (izquierda) la posicion, (derecha) el valor

def find_location(filas, element_to_find):
    '''Encuentra la ubicacion de una pieza en el rompecabezas.
       DEvuelve una tupla: fila, columna'''
    for ir, row in enumerate(filas):
        for ic, element in enumerate(row):
            if element == element_to_find:
                return ir, ic

def find_column_location(filas, element_to_find):
    for ir, row in enumerate(filas):
        for ic, element in enumerate(row):
            if element == element_to_find:
                return ic

#Posiciones unicamente de las filas en la se encuentras cada caracter similar
#se ordena sin importar la posicion de la columna, solo importa que el caracter
#en cuestion vuelva a su fila inical
posiciones_objetivo = {}
filas_objetivo = FINAL_STATE
i = 0
for numero in '12345abcde':
    posiciones_objetivo[numero] = find_location(filas_objetivo, numero)

    print(posiciones_objetivo)
    print(i)
    i += 1
    print(numero)

#Acciones en diagnoal tendra un menor coste
#Acciones en horizontal tendra un coste normal
#Acciones en vertical tendra un coste alto

class PuzzleProblem(ProblemaBusqueda):
    def acciones(self, estado):
        '''Devuelve una lista de piesas que se pueden mover a un espacio vacio.'''
        filas = estado
        #Esapcio vacio e1
        fila_a, columna_a = find_location(filas, 'a')
        #Espacio vacio e2
        fila_b, columna_b = find_location(filas, 'b')
        #Esapcio vacio e3
        fila_c, columna_c = find_location(filas, 'c')
        #Espacio vacio e4
        fila_d, columna_d = find_location(filas, 'd')
        #Espacio vacio e5
        fila_e, columna_e = find_location(filas, 'e')

        acciones = {}

        cantiadae = ['a','b','c','d','e']
        filas_espacios = [fila_a, fila_b, fila_c, fila_d, fila_e]
        columnas_espacios = [columna_a, columna_b, columna_c, columna_d, columna_e]

        acciones['a'] = []
        acciones['b'] = []
        acciones['c'] = []
        acciones['d'] = []
        acciones['e'] = []

        #Acciones de esquinas solo contienen 2 movimientos
        #Coste 2 moves
        def accion_esquinas_right_up(fila_e, columna_e, e):
            if fila_e == 0 and columna_e == 0:
                #move: left, down
                acciones[e].append(filas[fila_e][columna_e - 1])
                acciones[e].append(filas[fila_e + 1][columna_e])

        def accion_esqinas_left_up(fila_e, columna_e, e):
            if fila_e == 0 and columna_e == len(filas)-1:
                #move: right, down
                acciones[e].append(filas[fila_e][columna_e + 1])
                acciones[e].append(filas[fila_e + 1][columna_e])
            

        def accion_esquinas_right_down(fila_e, columna_e, e):
            if fila_e == len(filas) - 1 and columna_e == len(filas[0]) - 1:
                #move: up, left
                acciones[e].append(filas[fila_e - 1][columna_e])
                acciones[e].append(filas[fila_e][columna_e - 1])

        def accion_esqinas_left_down(fila_e, columna_e, e):
            if fila_e == len(filas) - 1 and columna_e == 0:
                #move: up, right
                acciones[e].append(filas[fila_e - 1][columna_e])
                acciones[e].append(filas[fila_e][columna_e + 1])
        
        #Acciones en muros sin contar con la esquinas
        #Coste de 3 movimientos
        def accion_wall_left(fila_e, columna_e, e):
            if fila_e >= 1 and fila_e <= len(filas[0]) - 2 and columna_e == 0:
                #move: right, up, down
                acciones[e].append(filas[fila_e][columna_e + 1])
                acciones[e].append(filas[fila_e - 1][columna_e])
                acciones[e].append(filas[fila_e + 1][columna_e])
        
        def accion_wall_right(fila_e, columna_e, e):
            if fila_e >= 1 and fila_e <= len(filas[0]) - 2 and columna_e == len(filas) - 1:
                #move: left, up, down
                acciones[e].append(filas[fila_e][columna_e - 1])
                acciones[e].append(filas[fila_e - 1][columna_e])
                acciones[e].append(filas[fila_e + 1][columna_e])

        def accion_wall_up(filas, fila_e, columna_e, e):
            if columna_e >= 1 and columna_e <= len(filas[0]) - 2 and fila_e == 0:
                #move: down, left, right
                acciones[e].append(filas[fila_e + 1][columna_e])
                acciones[e].append(filas[fila_e][columna_e - 1])
                acciones[e].append(filas[fila_e][columna_e + 1])

        def accion_wall_down(fila_e, columna_e, e):
             if columna_e >= 1 and columna_e <= len(filas[0]) - 2 and fila_e == len(filas) - 1:
                #move: up, left, right
                acciones[e].append(filas[fila_e - 1][columna_e])
                acciones[e].append(filas[fila_e][columna_e - 1])
                acciones[e].append(filas[fila_e][columna_e + 1])

        #Accion dentro de los muros
        #Coste de 4 movimeintos
        def accion_inside(fila_e, columna_e, e):
            if columna_e >= 2 and columna_e <= len(filas[0]) - 2 and fila_e >= 2 and fila_e <= len(filas) - 2:
                #move: up, down, left, right
                acciones[e].append(filas[fila_e - 1][columna_e])
                acciones[e].append(filas[fila_e + 1][columna_e])
                acciones[e].append(filas[fila_e][columna_e - 1])
                acciones[e].append(filas[fila_e][columna_e + 1])   

        #Acciones iteradas n veces por cada espacio vacio
        i = 0

        for e in cantiadae:

            accion_esquinas_right_up(filas_espacios[i], columnas_espacios[i], e)
            accion_esqinas_left_up(filas_espacios[i], columnas_espacios[i], e)
            accion_esquinas_right_down(filas_espacios[i], columnas_espacios[i], e)
            accion_esqinas_left_down(filas_espacios[i], columnas_espacios[i], e)

            accion_wall_left(filas_espacios[i], columnas_espacios[i], e)
            accion_wall_right(filas_espacios[i], columnas_espacios[i], e)
            accion_wall_down(filas_espacios[i], columnas_espacios[i], e)
            accion_wall_up(filas_espacios[i], columnas_espacios[i], e)

            accion_inside(filas_espacios[i], columnas_espacios[i], e)
            i += 1
        return acciones

    def resultado(self, estado, accion):
        '''Devuelve el resultado despues de mover una pieza a un espacio en vacio
        '''
        filas = estado

        fila_a, columna_a = find_location(filas, 'a')
        fila_n, columna_n = find_location(filas, accion)

        filas[fila_a][columna_a], filas[fila_n][columna_n] = filas[fila_n][columna_n], filas[fila_a][columna_a]

        fila_b, columna_b = find_location(filas, 'b')
        fila_n, columna_n = find_location(filas, accion)

        filas[fila_b][columna_b], filas[fila_n][columna_n] = filas[fila_n][columna_n], filas[fila_b][columna_b]

        fila_c, columna_c = find_location(filas, 'c')
        fila_n, columna_n = find_location(filas, accion)

        filas[fila_c][columna_c], filas[fila_n][columna_n] = filas[fila_n][columna_n], filas[fila_c][columna_c]

        fila_d, columna_d = find_location(filas, 'd')
        fila_n, columna_n = find_location(filas, accion)

        filas[fila_d][columna_d], filas[fila_n][columna_n] = filas[fila_n][columna_n], filas[fila_d][columna_d]
        
        fila_e, columna_e = find_location(filas, 'e')
        fila_n, columna_n = find_location(filas, accion)

        filas[fila_e][columna_e], filas[fila_n][columna_n] = filas[fila_n][columna_n], filas[fila_e][columna_e]

        return filas

    def es_objetivo(self, estado):
        '''Devuelve True si un estado es el estado_objetivo.'''
        return estado == FINAL_STATE

    #El coste va de este modo
    #Horizontales y Verticales costo similar
    def costo(self, estado1, accion, estado2):
        '''Devuelve el costo de ejecutar una accion. 
        '''

        return 1

    def heuristica(self, estado, accion):
        '''Devuelve una estimacion de la distancia
        de un estado a otro, utilizando la distancia manhattan.
        '''
        filas = estado
        #Tiene que ser un diccioanrio que represente la cantidad total de 
        #movimeintos que pueda realizar para generar la menor cantidad de estados
        acciones = {}
        acciones = accion

        acciona = acciones['a']
        accionb = acciones['b']
        accionc = acciones['c']
        acciond = acciones['d']
        accione = acciones['e']
        
        #aqui es donde ya no pude impelementear bien la heuristica, por el mal uso de 
        #Python, No he modificado nada desde la presentacion
        #Posiblemetne no de nada y no vaya a funcionar por la mala implementacion de mi
        #heuristica
        distancia = 0

        for numero in acciones['a']:
            
            fila_n, columna_n = find_location(filas, numero)
            fila_n_objetivo, col_n_goal = posiciones_objetivo[numero]

            distancia += abs(fila_n - fila_n_objetivo) + abs(columna_n - col_n_goal)

        for numero in acciones['b']:
            
            fila_n, columna_n = find_location(filas, numero)
            fila_n_objetivo, col_n_goal = posiciones_objetivo[numero]

            distancia += abs(fila_n - fila_n_objetivo) + abs(columna_n - col_n_goal)

        for numero in acciones['c']:
            
            fila_n, columna_n = find_location(filas, numero)
            fila_n_objetivo, col_n_goal = posiciones_objetivo[numero]

            distancia += abs(fila_n - fila_n_objetivo) + abs(columna_n - col_n_goal)

        for numero in acciones['d']:
            
            fila_n, columna_n = find_location(filas, numero)
            fila_n_objetivo, col_n_goal = posiciones_objetivo[numero]

            distancia += abs(fila_n - fila_n_objetivo) + abs(columna_n - col_n_goal)
        
        for numero in acciones['e']:
            
            fila_n, columna_n = find_location(filas, numero)
            fila_n_objetivo, col_n_goal = posiciones_objetivo[numero]

            distancia += abs(fila_n - fila_n_objetivo) + abs(columna_n - col_n_goal)

        return distancia

resultado = aestrella(PuzzleProblem(INITIAL_STATE))

print(resultado.camino())
print(resultado.acciones())

for accion, estado in resultado.camino():
    print('Move numero', accion)
    print(estado)
```
